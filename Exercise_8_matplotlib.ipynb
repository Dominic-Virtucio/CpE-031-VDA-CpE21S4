```python
import numpy as np
import matplotlib.pyplot as plt

g = np.array([35, 25, 25, 15])
grader = ["Grade 1=35%", "Grade 2=25%", "Grade 3=25%", "Grade 4=15%"]

# Define font styles
label_font = {'fontsize': 12, 'fontweight': 'bold', 'family': 'sans-serif'}
legend_font = {'fontsize': 10, 'fontweight': 'regular', 'family': 'sans-serif'}

# Create pie chart
plt.pie(g, labels=grader, textprops=label_font)

# Add legend
plt.legend(title="Grader", labels=grader, prop={'size': legend_font['fontsize'], 'weight': legend_font['fontweight']})

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.scatter(df['goal_difference'], df['rank'], color='blue')
plt.xlabel('Goal Difference')
plt.ylabel('Rank')
plt.title('Goal Difference vs. Rank')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['goals_scored'], df['rank'], color='blue')  # Simple blue color
plt.xlabel('Goals Scored')
plt.ylabel('Rank')
plt.title('Goals Scored vs. Rank')
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.scatter(df['goals_scored'], df['rank'], c=df['rank'], cmap='viridis')  # Color points by rank
plt.xlabel('Goals Scored')
plt.ylabel('Rank')
plt.title('Goals Scored vs. Rank')
plt.colorbar(label='Rank')  # Add colorbar to show rank mapping
plt.show()
```

```python
top_half = df[df['rank'] <= 12]['points']
bottom_half = df[df['rank'] > 12]['points']
plt.figure(figsize=(10, 6))
plt.hist(top_half, bins=10, alpha=0.5, label='Top Half')
plt.hist(bottom_half, bins=10, alpha=0.5, label='Bottom Half')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.title('Distribution of Points: Top Half vs. Bottom Half')
plt.legend(loc='upper right')
plt.show()
```

```python
top_5 = df[df['rank'] <= 5]
bottom_5 = df[df['rank'] > 20]
avg_goals_scored_top = top_5['goals_scored'].mean()
avg_goals_conceded_top = top_5['goals_conceded'].mean()
avg_goals_scored_bottom = bottom_5['goals_scored'].mean()
avg_goals_conceded_bottom = bottom_5['goals_conceded'].mean()

plt.figure(figsize=(10, 6))
width = 0.35
plt.bar(np.arange(2) - width/2, [avg_goals_scored_top, avg_goals_scored_bottom], width, label='Goals Scored', color='skyblue')
plt.bar(np.arange(2) + width/2, [avg_goals_conceded_top, avg_goals_conceded_bottom], width, label='Goals Conceded', color='coral')
plt.xticks(np.arange(2), ['Top 5', 'Bottom 5'])
plt.ylabel('Average Goals')
plt.title('Average Goals Scored and Conceded by Top 5 and Bottom 5 Teams')
plt.legend()
plt.show()
```
