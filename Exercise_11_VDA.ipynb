#Scatter plot using manual input

```python
#Manual Input
import numpy as np
x = np.array([0.5, 1, 1.3, 1.6, 2, 2.2, 2.5, 3.1, 3.9, 4.4])
y = np.array([1.9, 1.5, 1.2, 1, 0.8, 0.78, 0.65, 0.46, 0.3, 0.23])

slope, interceptr = np.polyfit(x,y,1)
y_pred = slope*x + interceptr

plt.scatter(x,y)
plt.plot(x,y_pred, color='red')
plt.show()
```

Scatter plot using uploaded csv file

```python
#Using csv upload
x = df['x']
y = df['y']

slope, intercept = np.polyfit(x, y, 1)

prediction = slope * x + intercept

#Plot the data points and the regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, prediction, color='red', label='Regression Line')
plt.show()
```

```python
x_values = df['x']
y_values = df['y']

slope, intercept = np.polyfit(x_values, y_values, 1)

predicted_values = slope * x_values + intercept

plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_values, color='blue', label='Data Points', alpha=0.7)
plt.plot(x_values, predicted_values, color='red', label='Regression Line', linewidth=2)

plt.title('Linear Regression of Data Points', fontsize=16)
plt.xlabel('X Values', fontsize=14)
plt.ylabel('Y Values', fontsize=14)
plt.legend()
plt.grid(True)

plt.show()
```

#Interpretation of data

- From my interpretation of the scatter plot, I can see that there is a negative correlation between the data points. This means that as the x-values increase, the y-values tend to decrease, which is clearly reflected in the graph. The red line represents a linear regression line, indicating a strong linear relationship between the data points. The points are closely clustered around this regression line, showing a clear trend of decreasing y-values as the x-values increase. This strengthens my observation of a strong negative correlation. Additionally, the Pearson r value of -0.9474635996 further confirms that there is indeed a very strong negative correlation between the variables.
