#What is data analytics?

Data analytics - this is the scientific process of transforming data into insights for making decision. This also involves collection of data, processing of data, organizing the collected data, and these outputs can be used to make outputs such as conclusion, making predictions. This is widely used for different purposes especially those that involves statistics like information of an individual, weather forecasts, business sales, and any other things where data is evident.

#Write a Pandas program to delete Data Frame row(s) based on given column value.

```python
#original data
import pandas as pd

data = {
    'col1':[1, 4, 3, 4, 5],
    'col2':[4, 5, 6, 7, 8],
    'col3':[7, 8, 9, 0, 1]
}

df = pd.DataFrame(data)

print(df)
```

```python
#new data
import pandas as pd

data = {
    'col1':[1, 4, 3, 4, 5],
    'col2':[4, 5, 6, 7, 8],
    'col3':[7, 8, 9, 0, 1]
}

df = pd.DataFrame(data)
df = df.drop([1])
print(df)
```
