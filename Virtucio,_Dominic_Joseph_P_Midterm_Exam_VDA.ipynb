#Questions
1: Is there a relationship between General Health and BMI

2: How does the sleeping hours on an individual affect their body weight

##For question 1: Is there a relationship between General Health and BMI

```python
import pandas as pd
import matplotlib.pyplot as plt

mean_bmi_by_health = df.groupby('GeneralHealth')['BMI'].mean()

plt.figure(figsize=(10, 6))
plt.bar(mean_bmi_by_health.index, mean_bmi_by_health.values)

plt.xlabel('General Health')
plt.ylabel('Average BMI')
plt.title('Average BMI by General Health')
plt.show()
```

###Question 1: Answer

- From the bar chart, I see that there isn’t a clear or consistent relationship between General Health and BMI. The average BMI values for each health category are quite similar, suggesting that people in better health don't necessarily have significantly different BMIs compared to those in poorer health. This could be due to individual variability, with BMI being influenced by many factors beyond just General Health, like genetics or lifestyle. It’s also possible that the data isn't comprehensive enough, or there may be other variables we haven’t considered. Overall, I can't conclude that there's a strong link between General Health and BMI without further analysis and additional data.


##For Question 2: How does the sleeping hoursHow does the sleeping hours on an individual affect their body weight

```python
import pandas as pd
import matplotlib.pyplot as plt

mean_bmi_by_health = df.groupby('SleepHours')['WeightInKilograms'].mean()

plt.figure(figsize=(10, 6))
plt.bar(mean_bmi_by_health.index, mean_bmi_by_health.values)

plt.xlabel('Sleep Hours')
plt.ylabel('Weight in Kilograms')
plt.title('How does the sleeping hours')
plt.legend
plt.show()
```

##Question 2: Answer
- In this bar graph regarding the question about whether there is a correlation in sleeping hours and wight of an individual, I employed a bar graph since appropriate for this data since it will immediately show the average weight for all sleep durations. The figure of the bar chart presented here indicates that there may not be a significant strong correlation between sleeping hours and weight, which suggests that merely sleep duration might not become a very important determinant to weight. As I am familiar with this kind of research on an individual's health because I conducted a study on the same subject in senior high school. I can tell you that an individual's average weight is not significantly influenced by how much sleep they get; instead, other factors like diet, exercise, and genetics can also affect an individual's total weight.
