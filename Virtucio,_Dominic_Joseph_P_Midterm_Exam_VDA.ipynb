##For question 1: Is there a relationship between General Health and BMI

```python
import pandas as pd
import matplotlib.pyplot as plt

mean_bmi_by_health = df.groupby('GeneralHealth')['BMI'].mean()

plt.figure(figsize=(10, 6))
plt.bar(mean_bmi_by_health.index, mean_bmi_by_health.values)

plt.xlabel('General Health')
plt.ylabel('Average BMI')
plt.title('Average BMI by General Health')
plt.show()
```

###Question 1: Answer

- From the bar chart, I see that there isnâ€™t a clear or consistent relationship between General Health and BMI. The average BMI values for each health category are quite similar, suggesting that people in better health don't necessarily have significantly different BMIs compared to those in poorer health. This could be due to individual variability, with BMI being influenced by many factors beyond just General Health, like genetics or lifestyle. Itâ€™s also possible that the data isn't comprehensive enough, or there may be other variables we havenâ€™t considered. Overall, I can't conclude that there's a strong link between General Health and BMI without further analysis and additional data.

##Questions

##Question 1: Is there a connection between an individual's smoking status and the likelihood of having Chronic Obstructive Pulmonary Disease (COPD)?

```python
copd_counts = df.groupby(['SmokerStatus', 'HadCOPD']).size().unstack(fill_value=0)

plt.figure(figsize=(25, 5))
copd_counts.plot(kind='bar', color=['skyblue', 'salmon'])

plt.xlabel('Smoking Status')
plt.ylabel('Number of Individuals')
plt.title('Number of Individuals with and without COPD by Smoking Status')

plt.xticks(rotation=30, ha='center')
plt.gca().set_xticklabels(['Current smoker - \nnow smokes every day', 'Current smoker - \nnow smokes some days', 'Former smoker\n', 'Never smoked\n'])

plt.legend(['No COPD', 'Had COPD'], title="COPD Status", loc='upper right')
plt.tight_layout()

plt.show()
```

###Question 1 Answer
- From the bar graph, there is a clear relationship between smoking and COPD. People who smoke daily have the most cases of COPD. This is a reflection of how smoking increases the risk. Former smokers also have a higher number of cases of COPD than never-smokers, which indicates that smoking can affect a person for some time even after quitting. The fewest cases were among those who never smoked, which shows that not smoking protects lung health. All things considered, the graph does support quitting smoking and the avoidance of smoking in life to decrease the risk of having lung disease.








##Question 2: Does the prevalence of difficulty walking increase with age, and if so, how does this relationship manifest across different age groups?

```python
walking_counts = df.groupby('AgeCategory')['DifficultyWalking'].value_counts().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
walking_counts.plot(kind='bar', color=['skyblue', 'salmon'])

plt.xlabel('Age Category')
plt.ylabel('Count of Individuals')
plt.title('Count of Individuals with Difficulty Walking by Age Category')
plt.xticks(rotation=45, ha='right')
plt.legend(['No Difficulty', 'Difficulty'], title="Difficulty Walking", loc='upper right')
plt.tight_layout()

plt.show()
```

###Question 2 - Answer
- The bar chart displays the connection between age and difficulties walking. I've found that people get worse at walking as they get older, especially those above the age of 65. This suggests that the elderly are more likely to have difficulty walking. However, this is not the true for everyone in the same age range; some have more obstacles than others. In general, this chart promotes the care and assistance of the elderly. It keeps people active and healthy, allowing them to live longer while walking or remaining independent.
