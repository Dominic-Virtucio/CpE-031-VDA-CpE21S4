##Analyze the data and generate the appropriate plot including its properties:

###a. Read each device's date of release and sales. Show it using a line plot.

####Unorganized data of sales and release dates

```python
plt.figure(figsize=(15, 6))
plt.plot(df['Date of Release'], df['Sales'], marker='o')

plt.xlabel('Date of Release', fontsize=14, fontweight='bold')
plt.ylabel('Sales', fontsize=14, fontweight='bold')
plt.title('Sales vs. Date of Release', fontsize=18, fontweight='bold')

plt.xticks(rotation=45)
plt.show()
```

###Analyze the data and generate the appropriate plot including its properties:
a. Read each device's date of release and sales. Show it using a line plot.

b. Get the total sales and show line plot with the following style properties.

Line Style dotted and Line-color should be red
Show legend at the lower right location.
X label name = Year
Y label name = Sales -Add a circle marker.
Line marker color as read
Line width should be 3

```python
plt.figure(figsize=(15, 6))
plt.plot(df['Date of Release'], df['Sales'], label='Sales', linestyle='dotted', color='red', marker='o', markerfacecolor='red', linewidth=3)

plt.xlabel('Year', fontsize=14, fontweight='bold')
plt.ylabel('Sales', fontsize=14, fontweight='bold')
plt.title('Sales vs. Date of Release', fontsize=18, fontweight='bold')

plt.xticks(rotation=45)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

###c. Read all product sales data and show it using a multiline plot. Label them.

```python
plt.figure(figsize=(15, 6))

for model in df['Model'].unique():
    model_data = df[df['Model'] == model].groupby('Date of Release')['Sales'].sum()
    plt.plot(model_data.index, model_data.values, marker='o', label=model)

plt.xlabel('Date of Release', fontsize=14, fontweight='bold')
plt.ylabel('Sales', fontsize=14, fontweight='bold')
plt.title('Sales for Each Product Model by Release Date', fontsize=18, fontweight='bold')
plt.legend(loc='upper left', fontsize=8)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.grid(True)
plt.tight_layout()
plt.show()
```

###d. Read each brand sales data and show it using the bar chart. Label them.

```python
plt.figure(figsize=(15, 6))

brand_sales = df.groupby('Brand')['Sales'].sum()
plt.bar(brand_sales.index, brand_sales.values, label='Sales')

plt.xlabel('Brand', fontsize=14, fontweight='bold')
plt.ylabel('Total Sales', fontsize=14, fontweight='bold')
plt.title('Total Sales for Each Brand', fontsize=18, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.legend(loc='upper left', fontsize=8)

plt.tight_layout()
plt.show()
```

###e. Read each device and show it using the histogram to see the most common Country of Origin. Label them.

```python
plt.figure(figsize=(15, 6))

df['Country of Origin'].value_counts().plot(kind='bar', color='hotpink', alpha=0.8, edgecolor='black')

plt.xlabel('Country of Origin', fontsize=14, fontweight='bold')
plt.ylabel('Frequency', fontsize=14, fontweight='bold')
plt.title('Frequency of Country of Origin for Devices', fontsize=18, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=8)

plt.tight_layout()
plt.show()
```

###f. Create a pie chart that shows each brand Country of Origin. Label them.

```python
brands_by_country = df.groupby('Country of Origin')['Brand'].nunique()

plt.figure(figsize=(8, 8))
brands_by_country.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors, explode=[0.1 if x == 'Taiwan' else 0 for x in brands_by_country.index])

plt.ylabel('')
plt.title('Brand Distribution by Country of Origin', fontsize=16, fontweight='bold')
plt.show()
```
